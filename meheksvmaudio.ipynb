```python
import numpy as np
import pandas as pd
from sklearn.model_selection import train_test_split
from xgboost import XGBClassifier
from sklearn.metrics import accuracy_score, classification_report
from sklearn.preprocessing import LabelEncoder
import os
import librosa
import matplotlib.pyplot as plt

# Define custom callback function to track training history
class HistoryCallback:
    def __init__(self):
        self.history = {'train_loss': [], 'val_loss': [], 'train_acc': [], 'val_acc': []}

    def __call__(self, env):
        self.history['train_loss'].append(env.evaluation_result_list[0][1])
        self.history['val_loss'].append(env.evaluation_result_list[1][1])
        self.history['train_acc'].append(env.evaluation_result_list[0][2])
        self.history['val_acc'].append(env.evaluation_result_list[1][2])

# Load the GTZAN dataset
gtzan_path = "/content/drive/MyDrive/Music Instrument/Indian Music Inst"
genres = os.listdir(gtzan_path)
labels = []
features = []

for genre in genres:
    genre_path = os.path.join(gtzan_path, genre)
    for filename in os.listdir(genre_path):
        file_path = os.path.join(genre_path, filename)
        try:
            y, sr = librosa.load(file_path, duration=30)  # Load audio file
            mfcc_feat = librosa.feature.mfcc(y=y, sr=sr)  # Extract MFCC features
            features.append(np.mean(mfcc_feat, axis=1))  # Take mean of MFCC features
            labels.append(genre)
        except Exception as e:
            print(f'Error processing {file_path}: {e}')

# Convert labels to numerical values
le = LabelEncoder()
y = le.fit_transform(labels)

# Split the dataset into train and test sets
X_train, X_test, y_train, y_test = train_test_split(features, y, test_size=0.2, random_state=42)

# Train XGBoost classifier with custom callback
xgb = XGBClassifier(n_estimators=100, random_state=42)
history = HistoryCallback()
xgb.fit(X_train, y_train, eval_set=[(X_train, y_train), (X_test, y_test)], eval_metric=['logloss', 'mlogloss', 'merror', 'mlogloss'], callbacks=[history])

# Predict on the test set
y_pred = xgb.predict(X_test)

# Convert predicted labels back to genre names
pred_genres = le.inverse_transform(y_pred)

# Calculate accuracy and print classification report
accuracy = accuracy_score(y_test, y_pred)
report = classification_report(y_test, y_pred, labels=np.arange(len(genres)), target_names=genres)
print(f'Accuracy: {accuracy}')
print(f'Classification Report:\n{report}')

# Plot training history
plt.plot(history.history['train_loss'], label='train_loss')
plt.plot(history.history['val_loss'], label='val_loss')
plt.plot(history.history['train_acc'], label='train_acc')
plt.plot(history.history['val_acc'], label='val_acc')
plt.xlabel('Epoch')
plt.legend()
plt.show()
```

```python
# RNN
import numpy as np
import pandas as pd
from sklearn.model_selection import train_test_split
from sklearn.preprocessing import LabelEncoder
from keras.models import Sequential
from keras.layers import Dense, LSTM
from sklearn.metrics import accuracy_score, classification_report
import os
import librosa

# Load the GTZAN dataset
gtzan_path = "/content/drive/MyDrive/Music Instrument/Indian Music Inst"
genres = os.listdir(gtzan_path)
labels = []
features = []

for genre in genres:
    genre_path = os.path.join(gtzan_path, genre)
    for filename in os.listdir(genre_path):
        file_path = os.path.join(genre_path, filename)
        try:
            y, sr = librosa.load(file_path, duration=30)  # Load audio file
            mfcc_feat = librosa.feature.mfcc(y=y, sr=sr)  # Extract MFCC features
            features.append(np.mean(mfcc_feat, axis=1))  # Take mean of MFCC features
            labels.append(genre)
        except Exception as e:
            print(f'Error processing {file_path}: {e}')

# Convert labels to numerical values
le = LabelEncoder()
y = le.fit_transform(labels)

# Split the dataset into train and test sets
X_train, X_test, y_train, y_test = train_test_split(features, y, test_size=0.2, random_state=42)

# Convert data to numpy arrays and reshape for RNN input
X_train = np.array(X_train)
X_train = np.reshape(X_train, (X_train.shape[0], 1, X_train.shape[1]))
X_test = np.array(X_test)
X_test = np.reshape(X_test, (X_test.shape[0], 1, X_test.shape[1]))

# Build the RNN model
model = Sequential()
model.add(LSTM(128, input_shape=(1, X_train.shape[2])))
model.add(Dense(len(genres), activation='softmax'))
model.compile(loss='sparse_categorical_crossentropy', optimizer='adam', metrics=['accuracy'])

# Train the RNN model
model.fit(X_train, y_train, epochs=50, batch_size=32)

# Predict on the test set
y_pred = model.predict(X_test)
y_pred = np.argmax(y_pred, axis=1)

# Convert predicted labels back to genre names
pred_genres = le.inverse_transform(y_pred)

# Calculate accuracy and print classification report
accuracy = accuracy_score(y_test, y_pred)
report = classification_report(y_test, y_pred, labels=np.arange(len(genres)), target_names=genres)
print(f'Accuracy: {accuracy}')
print(f'Classification Report:\n{report}')
```

```python


import numpy as np
from sklearn.model_selection import train_test_split
from sklearn.metrics import accuracy_score, classification_report
from sklearn.preprocessing import LabelEncoder
import os
import librosa
import torch
import torch.nn as nn
import torch.optim as optim
from torch.utils.data import Dataset, DataLoader

class AudioDataset(Dataset):
    def __init__(self, features, labels):
        self.features = features
        self.labels = labels

    def __len__(self):
        return len(self.features)

    def __getitem__(self, index):
        x = torch.Tensor(self.features[index])
        y = torch.Tensor([self.labels[index]])
        return x, y

class AudioCNN(nn.Module):
    def __init__(self, num_classes):
        super(AudioCNN, self).__init__()
        self.conv1 = nn.Conv2d(1, 32, kernel_size=3, stride=1)
        self.relu1 = nn.ReLU()
        self.maxpool1 = nn.MaxPool2d(kernel_size=2, stride=2)
        self.conv2 = nn.Conv2d(32, 64, kernel_size=3, stride=1)
        self.relu2 = nn.ReLU()
        self.maxpool2 = nn.MaxPool2d(kernel_size=2, stride=2)
        self.flatten = nn.Flatten()
        self.fc1 = nn.Linear(64 * 10 * 10, 128)
        self.relu3 = nn.ReLU()
        self.fc2 = nn.Linear(128, num_classes)

    def forward(self, x):
        x = self.conv1(x)
        x = self.relu1(x)
        x = self.maxpool1(x)
        x = self.conv2(x)
        x = self.relu2(x)
        x = self.maxpool2(x)
        x = self.flatten(x)
        x = self.fc1(x)
        x = self.relu3(x)
        x = self.fc2(x)
        return x



# Function to extract MFCC features from audio files
def extract_features(file_path):
    mfccs = []
    try:
        audio, _ = librosa.load(file_path)
        mfcc = librosa.feature.mfcc(y=audio, sr=22050, n_mfcc=20)
        mfccs.append(np.mean(mfcc, axis=1))
    except Exception as e:
        print(f'Error encountered while parsing file: {file_path}. Error: {e}')
        return None
    return np.array(mfccs)

# Load GTZAN dataset
data_path = "/content/drive/MyDrive/Music Instrument/Indian Music Inst"
genres = os.listdir(data_path)
all_features = []
all_labels = []
for genre in genres:
    genre_path = os.path.join(data_path, genre)
    for file in os.listdir(genre_path):
        file_path = os.path.join(genre_path, file)
        features = extract_features(file_path)
        if features is not None:
            all_features.append(features)
            all_labels.append(genre)

# Convert features and labels to numpy arrays
all_features = np.vstack(all_features)
all_labels = np.array(all_labels)

# Encode labels as integers
label_encoder = LabelEncoder()
all_labels_encoded = label_encoder.fit_transform(all_labels)

# Split dataset into training and testing sets
X_train, X_test, y_train, y_test = train_test_split(all_features, all_labels_encoded, test_size=0.2, random_state=42)

# Create AudioDataset and DataLoader
train_dataset = AudioDataset(X_train, y_train)
test_dataset = AudioDataset(X_test, y_test)
batch_size = 32
train_loader = DataLoader(train_dataset, batch_size=batch_size, shuffle=True)
test_loader = DataLoader(test_dataset, batch_size=batch_size, shuffle=False)

# Create the CNN model
num_classes = len(genres)
model = AudioCNN(num_classes)

# Define loss function and optimizer
criterion = nn.CrossEntropyLoss()
optimizer = optim.Adam(model.parameters(), lr=0.001)

# Training loop
num_epochs = 10
device = torch.device("cuda" if torch.cuda.is_available() else "cpu")
model.to(device)
for epoch in range(num_epochs):
    model.train()
    running_loss = 0.0
    for inputs, labels in train_loader:
        inputs = inputs.unsqueeze(1).to(device)  # Add a channel dimension
        labels = labels.squeeze().long().to(device)
        optimizer.zero_grad()
        outputs = model(inputs)
        loss = criterion(outputs, labels)
        loss.backward()
        optimizer.step()
        running_loss += loss.item() * inputs.size(0)
    epoch_loss = running_loss / len(train_dataset)
    print(f"Epoch {epoch+1}/{num_epochs}, Loss: {epoch_loss:.4f}")

# Evaluation on test set
model.eval()
predictions = []
true_labels = []
with torch.no_grad():
    for inputs, labels in test_loader:
        inputs = inputs.unsqueeze(1).to(device)  # Add a channel dimension
        labels = labels.squeeze().long().to(device)
        outputs = model(inputs)
        _, predicted = torch.max(outputs.data, 1)
        predictions.extend(predicted.cpu().numpy())
        true_labels.extend(labels.cpu().numpy())

# Decode integer labels back to original labels
predicted_labels = label_encoder.inverse_transform(predictions)
true_labels = label_encoder.inverse_transform(true_labels)

# Evaluate model performance
accuracy = accuracy_score(true_labels, predicted_labels)
report = classification_report(true_labels, predicted_labels, labels=genres)

# Print results
print(f'Accuracy: {accuracy}')
print(f'Classification Report:\n{report}')
```
